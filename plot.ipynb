```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# check current directory
import os
print(os.getcwd())
```

```python

# Plotting function for time to maturity
def plot_time_to_maturity():
    df = pd.read_csv('prices_to_plot/time_to_maturity.csv')
    plt.figure(figsize=(10, 6))
    plt.plot(df['TimeToMaturity'], df['OptionPrice'], marker='o')
    plt.title('Evolution of the call price with time to maturity')
    plt.xlabel('Time to Maturity (years)')
    plt.ylabel('Option Price ($)')
    plt.grid(True)
    plt.savefig('prices_to_plot/time_to_maturity.png')
    plt.show()

# Plotting function for volatility
def plot_volatility():
    df = pd.read_csv('prices_to_plot/volatility.csv')
    plt.figure(figsize=(10, 6))
    plt.plot(df['ATMVolatility'], df['OptionPrice'], marker='o')
    plt.title('Evolution of the call price with volatility')
    plt.xlabel('Annualized Volatility')
    plt.ylabel('Option Price ($)')
    plt.grid(True)
    plt.savefig('prices_to_plot/volatility.png')
    plt.show()

# Plotting function for time steps
def plot_time_steps():
    df = pd.read_csv('prices_to_plot/time_steps.csv')
    plt.figure(figsize=(10, 6))
    plt.plot(df['TimeSteps'], df['OptionPrice'], marker='o', label='Computed Option Price')
    ### plot real BS price
    plt.plot(df['TimeSteps'], df['BS_price'], marker='o', label='Theoretical BS Price')
    plt.title('Evolution of the call price with time steps')
    plt.xlabel('Number of Time Steps')
    plt.ylabel('Option Price ($)')
    plt.legend()
    plt.grid(True)
    plt.savefig('prices_to_plot/time_steps.png')
    plt.show()

# Plotting function for space steps
def plot_space_steps():
    df = pd.read_csv('prices_to_plot/space_steps.csv')
    plt.figure(figsize=(10, 6))
    plt.plot(df['SpaceSteps'], df['OptionPrice'], marker='o', label='Computed Option Price')
    ### plot real BS price
    plt.plot(df['SpaceSteps'], df['BS_price'], marker='o', label='Theoretical BS Price')
    plt.title('Evolution of the call price with space steps')
    plt.xlabel('Number of Space Steps')
    plt.ylabel('Option Price ($)')
    plt.legend()
    plt.grid(True)
    plt.savefig('prices_to_plot/space_steps.png')
    plt.show()

# Plotting function for MC and QMC prices
def plot_mc_qmc_prices(x_max):
    df = pd.read_csv('prices_to_plot/mc_qmc_prices.csv')
    plt.figure(figsize=(10, 6))
    plt.plot(df['NbSim'], df['MCPrice'], marker='o', label='MC Price')
    plt.plot(df['NbSim'], df['QMCPrice'], marker='o', label='QMC Price')
    plt.plot(df['NbSim'], df['BS_price'], marker='o', label='BS Price')
    plt.title('Call prices using MC & QMC methods')
    plt.xlabel('Number of Simulations')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    # plt.ylim(bottom=0)  # Set y-axis starting at 0 (optional)
    plt.xlim(0, x_max)  # Set x-axis range
    plt.savefig(f"prices_to_plot/mc_qmc_prices_{x_max}.png")
    plt.show()

```

```python

# Call the plotting functions
plot_time_to_maturity()
plot_volatility()
plot_time_steps()
plot_space_steps()
plot_mc_qmc_prices(100_000)
plot_mc_qmc_prices(11_000)
```
